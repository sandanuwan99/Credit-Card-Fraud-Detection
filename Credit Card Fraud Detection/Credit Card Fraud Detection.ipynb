```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
data = pd.read_csv('creditcard.csv')
data.head()
```

```python
# This dataset is highly unbaleanced
#0-->Nomal transaction
#1-->fraudulent transaction
```

```python
legit = data[data.Class==0]
fraud = data[data.Class==1]
```

```python
legit.shape
```

```python
fraud.shape
```

```python
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
# ommare the value for both transaction
```

```python
data.groupby('Class').mean()
```

```python
# build a sample dataset containg similar distribution and fraudulent transection
# number of fraudulent
```

```python
legit_sample = legit.sample(n=492)
```

```python
# concatenating two dataframes
```

```python
new_data = pd.concat([legit_sample,fraud],axis=0)
```

```python
new_data
```

```python
new_data['Class'].value_counts()
```

```python
new_data.groupby('Class').mean()
```

```python
# spliting the data into features and target
```

```python
x = new_data.drop(columns='Class',axis=1)
y = new_data['Class']
```

```python
x.head()
```

```python
y.head()
```

```python
#split the into training data and testing data
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
x.shape,x_train.shape,x_test.shape
```

```python
#model training 
#logistic regression
```

```python
model_lr = LogisticRegression()
```

```python
#training the logistic regression model with training data
```

```python
model_lr.fit(x_train,y_train)
```

```python
#Accuracy score
```

```python
x_train_prediction = model_lr.predict(x_train)
training_data_accuracy = accuracy_score(x_train_prediction,y_train)
print("Accuracy on traning data:",training_data_accuracy)
```

```python
x_test_predict = model_lr.predict(x_test)
testting_data_accuracy = accuracy_score(x_test_predict,y_test)
print('Accuracy on test data:',testting_data_accuracy)
```
